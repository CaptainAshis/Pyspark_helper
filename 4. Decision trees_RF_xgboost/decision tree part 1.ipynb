```
import pyspark
```

```
df=spark.sql("select * from college_6_csv")
```

```
df.printSchema()
```

```
from pyspark.ml.feature import VectorAssembler
```

```
df.columns
```

```
assembler=VectorAssembler(inputCols=['Apps','Accept','Enroll','Top10perc','Top25perc','F_Undergrad','P_Undergrad','Outstate','Room_Board','Books','Personal','PhD','Terminal','S_F_Ratio','perc_alumni','Expend','Grad_Rate'],outputCol='features')
```

```
output=assembler.transform(df)
output.show()
```

```
from pyspark.ml.feature import StringIndexer
```

```
indexer=StringIndexer(inputCol='Private',outputCol='PrivateIndex')
outputs_fixed=indexer.fit(output).transform(output)
outputs_fixed.printSchema()
```

```
final_data=outputs_fixed.select('features','PrivateIndex') 
```

```
train,test=final_data.randomSplit([0.7,0.3])
```

```
from pyspark.ml.classification import DecisionTreeClassifier,RandomForestClassifier,GBTClassifier
from pyspark.ml import Pipeline
```

```
dtc=DecisionTreeClassifier(featuresCol='features',labelCol='PrivateIndex')
rfc=RandomForestClassifier(numTrees=100,featuresCol='features',labelCol='PrivateIndex')
gbt=GBTClassifier(featuresCol='features',labelCol='PrivateIndex')
```

```
dtc_model=dtc.fit(train)
rfc_model=rfc.fit(train)
gbt_model=gbt.fit(train)
```

```
dtc_pred=dtc_model.transform(test)
rfc_pred=rfc_model.transform(test)
gbt_pred=gbt_model.transform(test)
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator,MulticlassClassificationEvaluator
```

```
binary_eval=BinaryClassificationEvaluator(labelCol='PrivateIndex')
print('DTC auc')
binary_eval.evaluate(dtc_pred)
```

```
binary_eval=BinaryClassificationEvaluator(labelCol='PrivateIndex')
print('RFC')
print(binary_eval.evaluate(rfc_pred))

binary_eval=BinaryClassificationEvaluator(labelCol='PrivateIndex')
print('GBT')
print(binary_eval.evaluate(gbt_pred))
```
